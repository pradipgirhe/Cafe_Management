```python
import tkinter as tk
import pandas as pd
import pywhatkit as pwk
from tkinter import messagebox

menu_data = {
    'Menu': ['Coffee', 'Tea', 'Samosa', 'Pizza', 'Vadapav'],
    'Price': [150, 20, 25, 175, 20]
}
menu = pd.DataFrame(menu_data)


quantities = [0] * len(menu['Menu'])

def update_quantity(index, change):
    quantities[index] += change
    quantities[index] = max(0, quantities[index])  # Ensure quantity doesn't go below 0
    quantity_labels[index].config(text=str(quantities[index]))
    calculate_total()

def calculate_total():
    total = 0
    for i, item in enumerate(menu['Menu']):
        total += quantities[i] * menu.loc[i, 'Price']
    total_label.config(text=f"Total: ₹{total}")
    return total

def reset():
    for i in range(len(quantities)):
        quantities[i] = 0
        quantity_labels[i].config(text="0")
    total_label.config(text="Total: ₹0")
    phone_entry.delete(0, tk.END)

def send_whatsapp_message():
    phone_number = phone_entry.get()
    if not phone_number:
        messagebox.showerror("Error", "Please enter a mobile number.")
        return
    total = calculate_total()
    if total == 0:
        messagebox.showerror("Error", "Order total is ₹0. Please add items to your order.")
        return
    
    message = f"Your order is confirmed. Your total bill is ₹{total}. Thank you for ordering!"
    
    try:
        pwk.sendwhatmsg_instantly(f"+91{phone_number}", message)
        messagebox.showinfo("Success", "Order confirmation sent via WhatsApp!")
    except Exception as e:
        messagebox.showerror("Error", f"Failed to send message: {str(e)}")

root = tk.Tk()
root.title("Restaurant Order System")
root.config(bg='black')

tk.Label(root, text="Welcome to Our Restaurant!", font=("Helvetica", 16), bg='white').grid(row=0, column=0, columnspan=3)
tk.Label(root, text="Menu", font=("Helvetica", 14), bg='white').grid(row=1, column=0)
tk.Label(root, text="Price", font=("Helvetica", 14), bg='white').grid(row=1, column=1)
tk.Label(root, text="Quantity", font=("Helvetica", 14), bg='white').grid(row=1, column=2)

quantity_labels = []
for i, item in enumerate(menu['Menu']):
    tk.Label(root, text=item).grid(row=i+2, column=0)
    tk.Label(root, text=f"₹{menu.loc[i, 'Price']}").grid(row=i+2, column=1)
    
    # Create - button
    minus_button = tk.Button(root, text="-", command=lambda i=i: update_quantity(i, -1), bg='lightgrey')
    minus_button.grid(row=i+2, column=2)
    
    # Label to display quantity
    quantity_label = tk.Label(root, text="0", width=5)
    quantity_label.grid(row=i+2, column=3)
    quantity_labels.append(quantity_label)
    
    # Create + button
    plus_button = tk.Button(root, text="+", command=lambda i=i: update_quantity(i, 1), bg='lightgrey')
    plus_button.grid(row=i+2, column=4)

# Phone number entry
tk.Label(root, text="Mobile Number:", font=("Helvetica", 14), bg='white').grid(row=len(menu['Menu'])+2, column=0)
phone_entry = tk.Entry(root)
phone_entry.grid(row=len(menu['Menu'])+2, column=1, columnspan=3)

tk.Button(root, text="Reset", command=reset, bg='pink').grid(row=len(menu['Menu'])+3, column=0, columnspan=2)
tk.Button(root, text="Calculate Total", command=calculate_total, bg='pink').grid(row=len(menu['Menu'])+3, column=2, columnspan=3)
tk.Button(root, text="Confirm Order", command=send_whatsapp_message, bg='green').grid(row=len(menu['Menu'])+4, column=0, columnspan=5)

total_label = tk.Label(root, text="Total: ₹0", font=("Helvetica", 14))
total_label.grid(row=len(menu['Menu'])+5, column=0, columnspan=5)

root.mainloop()
```
